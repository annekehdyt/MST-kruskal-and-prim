```python
import matplotlib.pyplot as plt
plt.plot([1,2,3,4])
plt.ylabel('some numbers')
plt.show()
```

```python
kruskal_time = [3623107, 42272, 47407, 51753, 65185, 64790, 87703, 95210]
prim_time = [89284, 101136, 109828, 108642, 118914, 178567, 233876, 371752]
vertices = [3,4,5,6,7,8,9,10]
```

```python
len(kruskal_time) == len(prim_time)
```

```python
plt.plot(vertices, prim_time, 'b', label='prim')
plt.plot(vertices, kruskal_time, 'r', label='kruskal')
plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)
plt.ylabel('Execute Time (ns)')
plt.xlabel('# of Vertex')
plt.title('Time comparison when |V| == |E|')
plt.show()
```

```python
plt.plot(vertices[1:], prim_time[1:], 'b', label='prim')
plt.plot(vertices[1:], kruskal_time[1:], 'r', label='kruskal')
plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)
plt.ylabel('Execute Time (ns)')
plt.xlabel('# of Vertex')
plt.title('Time comparison when |V| == |E|')
plt.show()
```

```python
kruskal_loop = [2,3,5,6,7,8,9,10]
prim_loop = [2,3,4,5,6,7,8,9]
```

```python
plt.plot(vertices, prim_loop, 'b', label='prim')
plt.plot(vertices, kruskal_loop, 'r', label='kruskal')
plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)
plt.ylabel('# of iteration')
plt.xlabel('# of Vertex')
plt.title('Iteration when |V| == |E|')
plt.show()
```
